```python
import pandas as pd
import matplotlib
import seaborn as sns
from sklearn.linear_model import PoissonRegressor, LinearRegression, GradientBoostingRegressor, RandomForestRegressor
import os
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Load your data (replace with your data loading method)
# data = pd.read_csv('./notebooks/Campylobacteriosis_case_rate_year_over_year.csv')
data = pd.read_csv('./Campylobacteriosis_case_rate_year_over_year.csv')
```

```python
data
```

```python


# data

# Define your features (X) and target variable (y)
X = data[["Year"]]
y = data["Cases"]


# Create a PoissonRegressor instance
poisson_model = PoissonRegressor()
linear_model = LinearRegression()
gradient_boost_model = GradientBoostingRegressor()
random_forest_model = RandomForestRegressor()


# Fit the model to your data
linear_model.fit(X, y)
poisson_model.fit(X, y)
gradient_boost_model.fit(X, y)
random_forest_model.fit(X, y)

# Make predictions
# predictions = model.predict(X)
# print(predictions)

next_year = np.array([[2024]])
linear_predicted_count = linear_model.predict(next_year)
poisson_predicted_count = poisson_model.predict(next_year)

print(f"Predicted count for 2025: {linear_predicted_count[0]:.2f}")
print(f"Predicted count for 2025: {poisson_predicted_count[0]:.2f}")

# Evaluate the model (optional)
# from sklearn.metrics import mean_squared_error
# mse = mean_squared_error(y, predictions)
# print('Mean Squared Error:', mse)
```

```python
# Plot the trend
plt.scatter(data['Year'], data['Cases'], color='blue', label='Actual')
plt.plot(data['Year'], poisson_model.predict(X), color='red', label='Poisson Regression')
plt.plot(data['Year'], linear_model.predict(X), color='red', label='Linear Line')
plt.scatter(2025, poisson_predicted_count, color='green', marker='x', s=200, label='Poisson Prediction for 2025')
plt.scatter(2025, linear_predicted_count, color='green', marker='x', s=200, label='Linear Prediction for 2025')
plt.xlabel('Year')
plt.ylabel('Cases')
plt.legend()
plt.show()
```
